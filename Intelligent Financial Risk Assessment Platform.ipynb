## 4. Data cleaning and pre processing

```python
def clean_emi_dataset(df):
    """
    Comprehensive cleaning function for EMI prediction dataset
    """
    df_clean = df.copy()
    
    # 1. Standardize Gender Column
    gender_mapping = {
        'Female': 'Female', 'female': 'Female', 'F': 'Female',
        'Male': 'Male', 'MALE': 'Male', 'M': 'Male', 'male': 'Male'
    }
    df_clean['gender'] = df_clean['gender'].map(gender_mapping).fillna('Unknown')
    
    # 2. Handle Missing Values in Categorical Columns
    categorical_columns = ['marital_status', 'education', 'employment_type', 'company_type', 'house_type']
    for col in categorical_columns:
        df_clean[col] = df_clean[col].fillna('Unknown')
        df_clean[col] = df_clean[col].replace('', 'Unknown')
    
    # 3. Clean Numeric Columns
    numeric_columns = [
        'monthly_salary', 'years_of_employment', 'monthly_rent', 
        'family_size', 'dependents', 'school_fees', 'college_fees',
        'travel_expenses', 'groceries_utilities', 'other_monthly_expenses',
        'current_emi_amount', 'credit_score', 'bank_balance', 'emergency_fund',
        'requested_amount', 'requested_tenure', 'max_monthly_emi'
    ]
    
    for col in numeric_columns:
        # Convert to numeric, coercing errors to NaN
        df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')
        # Fill missing values with median (more robust than mean)
        df_clean[col] = df_clean[col].fillna(df_clean[col].median())
        # Ensure no negative values for financial metrics
        if col in ['monthly_salary', 'bank_balance', 'emergency_fund']:
            df_clean[col] = df_clean[col].clip(lower=0)
    
    # 4. Clean 'existing_loans' column
    df_clean['existing_loans'] = df_clean['existing_loans'].map({
        'Yes': 'Yes', 'No': 'No'
    }).fillna('No')
    
    # 5. Handle Target Variable
    df_clean['emi_eligibility'] = df_clean['emi_eligibility'].map({
        'Eligible': 'Eligible', 
        'Not_Eligible': 'Not_Eligible', 
        'High_Risk': 'High_Risk'
    }).fillna('Not_Eligible')
    
    # 6. Create new features for better modeling
    df_clean['disposable_income'] = (
        df_clean['monthly_salary'] - 
        df_clean['monthly_rent'] - 
        df_clean['school_fees'] - 
        df_clean['college_fees'] -
        df_clean['travel_expenses'] -
        df_clean['groceries_utilities'] -
        df_clean['other_monthly_expenses'] -
        df_clean['current_emi_amount']
    )
    
    df_clean['debt_to_income_ratio'] = (
        df_clean['current_emi_amount'] / 
        (df_clean['monthly_salary'] + 1)  # +1 to avoid division by zero
    )
    
    df_clean['savings_ratio'] = (
        df_clean['bank_balance'] / 
        (df_clean['monthly_salary'] + 1)
    )
    
    # Fix age column - handle invalid values like "58.0.0"
    df_clean['age'] = pd.to_numeric(df_clean['age'], errors='coerce')
    
    # Fill missing ages with median
    df_clean['age'] = df_clean['age'].fillna(df_clean['age'].median())
    
    # Ensure age is integer and reasonable
    df_clean['age'] = df_clean['age'].astype(int)
    df_clean['age'] = df_clean['age'].clip(18, 80)  # Reasonable age range
    
    # 7. Remove duplicates if any
    df_clean = df_clean.drop_duplicates()
    
    # 8. Reset index
    df_clean = df_clean.reset_index(drop=True)
    
    return df_clean

# Apply cleaning function
df_cleaned = clean_emi_dataset(df)
```

```python
def create_advanced_features(df):
    """
    Create advanced features for better predictive power
    """
    df_fe = df.copy()
    
    # A. Financial Health Features
    print("Creating Financial Health Features...")
    
    # 1. Expense Ratios
    df_fe['expense_to_income_ratio'] = (
        (df_fe['monthly_rent'] + df_fe['school_fees'] + df_fe['college_fees'] + 
         df_fe['travel_expenses'] + df_fe['groceries_utilities'] + df_fe['other_monthly_expenses']) / 
        (df_fe['monthly_salary'] + 1)
    )
    
    # 2. Savings and Investment Capacity
    df_fe['savings_capacity'] = (
        (df_fe['monthly_salary'] - 
         df_fe['monthly_rent'] - 
         df_fe['school_fees'] - 
         df_fe['college_fees'] -
         df_fe['travel_expenses'] -
         df_fe['groceries_utilities'] -
         df_fe['other_monthly_expenses'] -
         df_fe['current_emi_amount']) / 
        (df_fe['monthly_salary'] + 1)
    )
    
    # 3. Financial Stability Score (0-100)
    df_fe['financial_stability_score'] = (
        (df_fe['credit_score'] / 10) +  # Credit score component (0-100)
        (np.minimum(df_fe['savings_ratio'] * 20, 30)) +  # Savings component (0-30)
        (np.minimum((1 - df_fe['debt_to_income_ratio']) * 30, 30)) +  # Debt management (0-30)
        (np.minimum(df_fe['years_of_employment'] * 2, 20))  # Employment stability (0-20)
    )
    
    # 4. Loan Affordability Metrics
    df_fe['emi_affordability_ratio'] = (
        df_fe['max_monthly_emi'] / 
        (df_fe['monthly_salary'] - df_fe['current_emi_amount'] + 1)
    )
    
    df_fe['requested_loan_affordability'] = (
        (df_fe['requested_amount'] / df_fe['requested_tenure']) /  # Approx monthly payment
        (df_fe['monthly_salary'] - df_fe['current_emi_amount'] + 1)
    )
    
    # B. Demographic and Behavioral Features
    print("Creating Demographic and Behavioral Features...")
    
    # 5. Life Stage Classification
    conditions = [
        (df_fe['age'] < 30) & (df_fe['marital_status'] == 'Single'),
        (df_fe['age'] < 30) & (df_fe['marital_status'] == 'Married'),
        (df_fe['age'] < 40) & (df_fe['dependents'] > 0),
        (df_fe['age'] >= 40) & (df_fe['dependents'] > 0),
        (df_fe['age'] >= 50)
    ]
    choices = ['Young_Single', 'Young_Married', 'Family_Stage', 'Established_Family', 'Senior']
    df_fe['life_stage'] = np.select(conditions, choices, default='Other')
    
    # 6. Employment Quality Score
    employment_score_map = {
        'Government': 10,
        'Private': 8,
        'Self-employed': 6,
        'Other': 5
    }
    company_size_map = {
        'MNC': 10,
        'Large Indian': 9,
        'Mid-size': 7,
        'Startup': 5,
        'Small': 4
    }
    
    df_fe['employment_quality_score'] = (
        df_fe['employment_type'].map(employment_score_map) +
        df_fe['company_type'].map(company_size_map) +
        np.minimum(df_fe['years_of_employment'] * 2, 20)  # Max 20 points for tenure
    )
    
    # C. Loan-specific Features
    print("Creating Loan-specific Features...")
    
    # 7. Loan Purpose Encoding
    loan_purpose_risk = {
        'Education EMI': 3,      # Lower risk - investment in future
        'Home Appliances EMI': 5, # Medium risk - depreciating assets
        'Vehicle EMI': 6,        # Medium-high risk - depreciating but essential
        'Personal Loan EMI': 8,   # Higher risk - unspecified purpose
        'E-commerce Shopping EMI': 9  # Highest risk - discretionary spending
    }
    df_fe['loan_purpose_risk_score'] = df_fe['emi_scenario'].map(loan_purpose_risk)
    
    # 8. Loan-to-Value Equivalent
    df_fe['loan_to_asset_ratio'] = (
        df_fe['requested_amount'] / 
        (df_fe['bank_balance'] + df_fe['emergency_fund'] + 1)
    )
    
    # D. Risk Assessment Features
    print("Creating Risk Assessment Features...")
    
    # 9. Composite Risk Score
    df_fe['composite_risk_score'] = (
        (df_fe['debt_to_income_ratio'] * 25) +
        ((1 - (df_fe['credit_score'] / 1000)) * 30) +
        (df_fe['loan_purpose_risk_score'] * 15) +
        (df_fe['expense_to_income_ratio'] * 20) +
        ((1 - df_fe['savings_capacity']) * 10)
    )
    
    # 10. Emergency Fund Adequacy
    df_fe['emergency_fund_adequacy'] = (
        df_fe['emergency_fund'] / 
        (df_fe['monthly_rent'] + df_fe['groceries_utilities'] + df_fe['other_monthly_expenses'])
    )
    
    # E. Interaction Features
    print("Creating Interaction Features...")
    
    # 11. High Debt High Income
    df_fe['high_debt_high_income'] = (
        (df_fe['debt_to_income_ratio'] > 0.4) & 
        (df_fe['monthly_salary'] > df_fe['monthly_salary'].median())
    ).astype(int)
    
    # 12. Good Credit Low Income
    df_fe['good_credit_low_income'] = (
        (df_fe['credit_score'] > 700) & 
        (df_fe['monthly_salary'] < df_fe['monthly_salary'].median())
    ).astype(int)
    
    # 13. Family Financial Pressure
    df_fe['family_financial_pressure'] = (
        df_fe['dependents'] * df_fe['expense_to_income_ratio']
    )
    
    return df_fe

# Apply advanced feature engineering
df_engineered = create_advanced_features(df_engineered)
print(f"Dataset shape after feature engineering: {df_engineered.shape}")
```

### Feature Importance Analysis

```python
def analyze_feature_importance(X, y, numerical_features):
    """
    Analyze feature importance using multiple methods
    """
    from sklearn.ensemble import RandomForestClassifier
    from sklearn.feature_selection import SelectKBest, f_classif, mutual_info_classif
    
    # Convert categorical target to numerical for feature selection
    label_encoder = LabelEncoder()
    y_encoded = label_encoder.fit_transform(y)
    
    # Use only numerical features for importance analysis
    X_numerical = X[numerical_features].fillna(0)
    
    print("Performing Feature Importance Analysis...")
    
    # Method 1: Random Forest Feature Importance
    rf = RandomForestClassifier(n_estimators=100, random_state=42)
    rf.fit(X_numerical, y_encoded)
    rf_importance = pd.DataFrame({
        'feature': numerical_features,
        'rf_importance': rf.feature_importances_
    }).sort_values('rf_importance', ascending=False)
    
    # Method 2: ANOVA F-value
    selector_f = SelectKBest(score_func=f_classif, k='all')
    selector_f.fit(X_numerical, y_encoded)
    f_scores = pd.DataFrame({
        'feature': numerical_features,
        'f_score': selector_f.scores_
    }).sort_values('f_score', ascending=False)
    
    # Method 3: Mutual Information
    selector_mi = SelectKBest(score_func=mutual_info_classif, k='all')
    selector_mi.fit(X_numerical, y_encoded)
    mi_scores = pd.DataFrame({
        'feature': numerical_features,
        'mi_score': selector_mi.scores_
    }).sort_values('mi_score', ascending=False)
    
    # Combine results
    importance_df = rf_importance.merge(f_scores, on='feature').merge(mi_scores, on='feature')
    importance_df['composite_score'] = (
        importance_df['rf_importance'] + 
        importance_df['f_score'] / importance_df['f_score'].max() + 
        importance_df['mi_score'] / importance_df['mi_score'].max()
    )
    importance_df = importance_df.sort_values('composite_score', ascending=False)
    
    return importance_df

# Perform feature importance analysis
importance_df = analyze_feature_importance(X, y, numerical_features)

print("\nTop 20 Most Important Features:")
print(importance_df.head(20).round(4))
```

### Data Preprocessing Pipeline

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

def create_preprocessing_pipeline(categorical_features, numerical_features, top_k_features=30):
    """
    Create a comprehensive preprocessing pipeline
    """
    
    # Select top features based on importance analysis
    top_features = importance_df.head(top_k_features)['feature'].tolist()
    
    # Update feature lists to include only top features
    numerical_features_top = [f for f in numerical_features if f in top_features]
    categorical_features_top = [f for f in categorical_features if f in top_features]
    
    print(f"Selected {len(numerical_features_top)} numerical features")
    print(f"Selected {len(categorical_features_top)} categorical features")
    
    # Numerical preprocessing pipeline
    numerical_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler())
    ])
    
    # Categorical preprocessing pipeline
    categorical_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
        ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
    ])
    
    # Combine preprocessing steps
    preprocessor = ColumnTransformer(
        transformers=[
            ('num', numerical_transformer, numerical_features_top),
            ('cat', categorical_transformer, categorical_features_top)
        ]
    )
    
    return preprocessor, numerical_features_top, categorical_features_top

# Create preprocessing pipeline
preprocessor, num_features, cat_features = create_preprocessing_pipeline(
    categorical_features, numerical_features, top_k_features=30
)
```

###  Model 1: Logistic Regression 

```python
def train_logistic_regression(X_train, X_test, y_train, y_test, class_names):
    """
    Train and evaluate Logistic Regression model
    """
    print("="*60)
    print("MODEL 1: LOGISTIC REGRESSION")
    print("="*60)
    
    # Create pipeline with scaling
    lr_pipeline = Pipeline([
        ('scaler', StandardScaler()),
        ('lr', LogisticRegression(
            multi_class='multinomial',
            solver='lbfgs',
            max_iter=1000,
            random_state=42,
            class_weight='balanced'
        ))
    ])
    
    # Train model
    lr_pipeline.fit(X_train, y_train)
    
    # Predictions
    y_pred_lr = lr_pipeline.predict(X_test)
    y_pred_proba_lr = lr_pipeline.predict_proba(X_test)
    
    # Evaluate model
    accuracy_lr = accuracy_score(y_test, y_pred_lr)
    f1_lr = f1_score(y_test, y_pred_lr, average='weighted')
    
    print(f"Accuracy: {accuracy_lr:.4f}")
    print(f"F1-Score: {f1_lr:.4f}")
    
    # Cross-validation
    cv_scores = cross_val_score(lr_pipeline, X_train, y_train, cv=5, scoring='accuracy')
    print(f"Cross-validation Accuracy: {cv_scores.mean():.4f} (+/- {cv_scores.std() * 2:.4f})")
    
    # Feature coefficients (for interpretability)
    feature_importance = pd.DataFrame({
        'feature': X_train.columns,
        'coefficient': lr_pipeline.named_steps['lr'].coef_[0]  # Using first class coefficients
    }).sort_values('coefficient', key=abs, ascending=False)
    
    print("\nTop 10 Most Important Features (by coefficient magnitude):")
    print(feature_importance.head(10))
    
    return lr_pipeline, y_pred_lr, y_pred_proba_lr, accuracy_lr, f1_lr

# Train Logistic Regression
lr_model, y_pred_lr, y_pred_proba_lr, acc_lr, f1_lr = train_logistic_regression(
    X_train_clean, X_test_clean, y_train_enc, y_test_enc, class_names
)
```

### Data preparation for regression

```python
def prepare_regression_data(df, target_col='max_monthly_emi', test_size=0.2, random_state=42):
    """
    Prepare data for regression task
    """
    # Separate features and target
    X = df.drop([target_col, 'emi_eligibility'], axis=1)  # Remove target and classification target
    y = df[target_col]
    
    # Handle categorical variables
    categorical_cols = X.select_dtypes(include=['object']).columns
    numerical_cols = X.select_dtypes(include=[np.number]).columns
    
    print(f"Categorical columns: {len(categorical_cols)}")
    print(f"Numerical columns: {len(numerical_cols)}")
    
    # One-hot encode categorical variables
    if len(categorical_cols) > 0:
        X_encoded = pd.get_dummies(X, columns=categorical_cols, drop_first=True)
    else:
        X_encoded = X.copy()
    
    print(f"Features after encoding: {X_encoded.shape[1]}")
    
    # Split data
    X_train, X_test, y_train, y_test = train_test_split(
        X_encoded, y, 
        test_size=test_size, 
        random_state=random_state
    )
    
    print(f"Training set: {X_train.shape}")
    print(f"Test set: {X_test.shape}")
    
    # Target statistics
    print(f"\nTarget variable statistics:")
    print(f"Training - Mean: {y_train.mean():.2f}, Std: {y_train.std():.2f}")
    print(f"Test - Mean: {y_test.mean():.2f}, Std: {y_test.std():.2f}")
    
    return X_train, X_test, y_train, y_test, X_encoded.columns

# Prepare regression data
X_train_reg, X_test_reg, y_train_reg, y_test_reg, feature_names = prepare_regression_data(df_final)
```

###  Feature Selection for Regression

```python
def select_regression_features(X_train, y_train, X_test, k=30):
    """
    Select most relevant features for regression
    """
    # Feature selection using F-regression
    selector = SelectKBest(score_func=f_regression, k=k)
    X_train_selected = selector.fit_transform(X_train, y_train)
    X_test_selected = selector.transform(X_test)
    
    # Get selected feature names
    selected_mask = selector.get_support()
    selected_features = X_train.columns[selected_mask]
    
    print(f"Selected {len(selected_features)} features for regression")
    print("\nTop selected features:")
    scores = selector.scores_[selected_mask]
    feature_scores = pd.DataFrame({
        'feature': selected_features,
        'f_score': scores
    }).sort_values('f_score', ascending=False)
    
    print(feature_scores.head(15))
    
    # Plot feature scores
    plt.figure(figsize=(12, 8))
    top_features = feature_scores.head(15)
    plt.barh(top_features['feature'], top_features['f_score'])
    plt.xlabel('F-Score')
    plt.title('Top 15 Features for EMI Amount Prediction')
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.show()
    
    return X_train_selected, X_test_selected, selected_features, feature_scores

# Select features for regression
X_train_sel, X_test_sel, selected_features, feature_scores = select_regression_features(
    X_train_reg, y_train_reg, X_test_reg, k=25
)
```

```python
def train_linear_regression(X_train, X_test, y_train, y_test, feature_names):
    """
    Train and evaluate Linear Regression model
    """
    print("="*60)
    print("MODEL 1: LINEAR REGRESSION")
    print("="*60)
    
    # Create pipeline with scaling
    lr_pipeline = Pipeline([
        ('scaler', StandardScaler()),
        ('lr', LinearRegression())
    ])
    
    # Train model
    lr_pipeline.fit(X_train, y_train)
    
    # Predictions
    y_pred_lr = lr_pipeline.predict(X_test)
    
    # Evaluate model
    mse_lr = mean_squared_error(y_test, y_pred_lr)
    mae_lr = mean_absolute_error(y_test, y_pred_lr)
    r2_lr = r2_score(y_test, y_pred_lr)
    rmse_lr = np.sqrt(mse_lr)
    
    print(f"RÂ² Score: {r2_lr:.4f}")
    print(f"RMSE: {rmse_lr:.2f}")
    print(f"MAE: {mae_lr:.2f}")
    print(f"MSE: {mse_lr:.2f}")
    
    # Cross-validation
    cv_scores = cross_val_score(lr_pipeline, X_train, y_train, cv=5, scoring='r2')
    print(f"Cross-validation RÂ²: {cv_scores.mean():.4f} (+/- {cv_scores.std() * 2:.4f})")
    
    # Feature coefficients
    if hasattr(lr_pipeline.named_steps['lr'], 'coef_'):
        coefficients = lr_pipeline.named_steps['lr'].coef_
        feature_importance_lr = pd.DataFrame({
            'feature': feature_names,
            'coefficient': coefficients
        }).sort_values('coefficient', key=abs, ascending=False)
        
        print("\nTop 10 Most Influential Features (by coefficient magnitude):")
        print(feature_importance_lr.head(10))
    else:
        feature_importance_lr = None
    
    return lr_pipeline, y_pred_lr, mse_lr, mae_lr, r2_lr, rmse_lr, feature_importance_lr

# Train Linear Regression
lr_model, y_pred_lr, mse_lr, mae_lr, r2_lr, rmse_lr, feature_importance_lr = train_linear_regression(
    X_train_sel, X_test_sel, y_train_reg, y_test_reg, selected_features
)
```

```python
import streamlit as st
import pandas as pd
import numpy as np
import pickle
import joblib
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# ------------------------------
# Page configuration
# ------------------------------
st.set_page_config(
    page_title="EMI Amount Predictor",
    page_icon="ðŸ’°",
    layout="wide",
    initial_sidebar_state="expanded"
)

# ------------------------------
# Custom CSS for better styling
# ------------------------------
st.markdown(
    """
<style>
    .main-header { font-size: 3rem; color: #1f77b4; text-align: center; margin-bottom: 2rem; }
    .prediction-box { background-color: #f0f2f6; padding: 2rem; border-radius: 10px; border-left: 5px solid #1f77b4; margin: 1rem 0; }
    .feature-importance { background-color: #e8f4f8; padding: 1rem; border-radius: 5px; margin: 0.5rem 0; }
    .stButton button { background-color: #1f77b4; color: white; font-weight: bold; border-radius: 5px; padding: 0.5rem 2rem; }
</style>
""",
    unsafe_allow_html=True,
)

# ------------------------------
# Utility: one-hot helper that aligns to saved feature_names
# ------------------------------
CATEGORICAL_FIELDS = [
    "marital_status",
    "education",
    "employment_type",
    "company_type",
    "emi_scenario",
    "existing_loans"
]

NUMERIC_FIELDS = [
    "age",
    "monthly_salary",
    "years_of_employment",
    "monthly_rent",
    "family_size",
    "dependents",
    "school_fees",
    "college_fees",
    "travel_expenses",
    "groceries_utilities",
    "other_monthly_expenses",
    "current_emi_amount",
    "credit_score",
    "bank_balance",
    "emergency_fund",
    "requested_amount",
    "requested_tenure",
]


def build_row_aligned_to_features(input_data: dict, feature_names: list[str]) -> pd.DataFrame:
    """Create a single-row dataframe with columns exactly equal to feature_names.
    - Numeric features are filled directly from input_data (0 default).
    - Categorical features are mapped to one-hot columns using the pattern f"{field}_{value}".
      Only columns that exist in feature_names are set to 1; all others remain 0.
    This mirrors the training pipeline that produced 'feature_names'.
    """
    # Start with all zeros
    row = pd.Series(0, index=feature_names, dtype=float)

    # Fill numeric fields if present in feature_names
    for f in NUMERIC_FIELDS:
        if f in feature_names:
            row[f] = float(input_data.get(f, 0) or 0)

    # Handle categoricals via one-hot using saved feature names
    for f in CATEGORICAL_FIELDS:
        val = input_data.get(f, None)
        if val is None:
            continue
        # Expected column name pattern in training: e.g., "marital_status_Single"
        candidate_col = f"{f}_{val}"
        if candidate_col in feature_names:
            row[candidate_col] = 1.0
        else:
            # If the exact value column isn't present, try a case-normalized match
            # (useful when training used different capitalization)
            lowered = {c.lower(): c for c in feature_names if c.startswith(f + "_")}
            key = candidate_col.lower()
            if key in lowered:
                row[lowered[key]] = 1.0
            # Otherwise leave all zeros for this group (unseen category at inference)

    return pd.DataFrame([row.values], columns=feature_names)


# ------------------------------
# EMI Predictor wrapper
# ------------------------------
class EMIPredictor:
    def __init__(self):
        self.model = None
        self.scaler: StandardScaler | None = None
        self.feature_names: list[str] | None = None
        self.load_model()

    def load_model(self):
        """Load the trained model and preprocessing objects."""
        try:
            self.model = joblib.load("best_emi_amount_model_xgboost_regressor.pkl")

            with open("model_features.pkl", "rb") as f:
                model_info = pickle.load(f)
                self.feature_names = model_info.get("feature_names")
                self.scaler = model_info.get("scaler")

            if not self.feature_names:
                st.error("Model metadata missing 'feature_names'. Ensure training exported them.")
                st.stop()

            st.sidebar.success("âœ… Model loaded successfully!")
        except FileNotFoundError:
            st.error("âŒ Model files not found. Place the .pkl files in the working directory.")
            st.stop()
        except Exception as e:
            st.error(f"âŒ Failed to load model: {e}")
            st.stop()

    def preprocess_input(self, input_data: dict):
        """Preprocess user input to match model requirements (aligned, scaled)."""
        try:
            assert self.feature_names is not None
            # 1) Align to training features with one-hot handling
            df_row = build_row_aligned_to_features(input_data, self.feature_names)

            # 2) Scale if scaler exists
            if self.scaler is not None:
                arr = self.scaler.transform(df_row)
            else:
                arr = df_row.values

            return arr
        except Exception as e:
            st.error(f"Error in preprocessing: {e}")
            return None

    def predict_emi(self, input_data: dict):
        """Predict EMI amount based on input data."""
        try:
            processed = self.preprocess_input(input_data)
            if processed is None:
                return None
            pred = float(self.model.predict(processed)[0])
            return max(0.0, pred)
        except Exception as e:
            st.error(f"Error in prediction: {e}")
            return None


# ------------------------------
# UI sections
# ------------------------------

def main():
    predictor = EMIPredictor()

    st.markdown('<h1 class="main-header">ðŸ’° EMI Amount Predictor</h1>', unsafe_allow_html=True)
    st.markdown(
        """
        This application predicts the maximum affordable EMI amount based on your financial profile,
        employment details, and credit history. Fill in the form below to get your personalized EMI prediction.
        """
    )

    tab1, tab2, tab3, tab4 = st.tabs(["ðŸ“Š EMI Prediction", "ðŸ“ˆ Financial Analysis", "â„¹ï¸ About", "ðŸ”§ Model Info"])

    with tab1:
        display_prediction_form(predictor)
    with tab2:
        display_financial_analysis()
    with tab3:
        display_about_section()
    with tab4:
        display_model_info(predictor)


# ------------------------------
# Prediction form
# ------------------------------

def display_prediction_form(predictor: EMIPredictor):
    col1, col2 = st.columns([2, 1])

    with col1:
        st.subheader("Personal & Financial Information")

        # Personal Information
        with st.expander("ðŸ‘¤ Personal Details", expanded=True):
            col1a, col1b = st.columns(2)
            with col1a:
                age = st.slider("Age", min_value=18, max_value=70, value=35, help="Your current age")
                marital_status = st.selectbox("Marital Status", ["Single", "Married", "Divorced", "Widowed"])
                education = st.selectbox("Education Level", [
                    "High School",
                    "Graduate",
                    "Post Graduate",
                    "Professional",
                    "Other",
                ])
            with col1b:
                family_size = st.slider("Family Size", min_value=1, max_value=10, value=3)
                dependents = st.slider("Number of Dependents", min_value=0, max_value=8, value=1)

        # Employment Information
        with st.expander("ðŸ’¼ Employment Details", expanded=True):
            col2a, col2b = st.columns(2)
            with col2a:
                monthly_salary = st.number_input(
                    "Monthly Salary (â‚¹)", min_value=0, max_value=500000, value=50000, step=5000,
                    help="Your gross monthly income",
                )
                employment_type = st.selectbox("Employment Type", ["Private", "Government", "Self-employed", "Other"])
                years_of_employment = st.slider("Years of Employment", min_value=0.0, max_value=40.0, value=5.0, step=0.5)
            with col2b:
                company_type = st.selectbox("Company Type", ["MNC", "Large Indian", "Mid-size", "Startup", "Small", "Other"])

        # Financial Information
        with st.expander("ðŸ’° Financial Details", expanded=True):
            col3a, col3b = st.columns(2)
            with col3a:
                credit_score = st.slider("Credit Score", min_value=300, max_value=900, value=650, help="Your credit score (300-900)")
                bank_balance = st.number_input("Bank Balance (â‚¹)", min_value=0, max_value=2000000, value=100000, step=10000)
                emergency_fund = st.number_input("Emergency Fund (â‚¹)", min_value=0, max_value=1000000, value=50000, step=5000)
            with col3b:
                monthly_rent = st.number_input("Monthly Rent (â‚¹)", min_value=0, max_value=100000, value=15000, step=1000)
                current_emi_amount = st.number_input("Current EMI Obligations (â‚¹)", min_value=0, max_value=100000, value=0, step=1000)

        # Expenses
        with st.expander("ðŸ“Š Monthly Expenses", expanded=False):
            col4a, col4b = st.columns(2)
            with col4a:
                school_fees = st.number_input("School Fees (â‚¹)", min_value=0, max_value=50000, value=0)
                college_fees = st.number_input("College Fees (â‚¹)", min_value=0, max_value=100000, value=0)
                travel_expenses = st.number_input("Travel Expenses (â‚¹)", min_value=0, max_value=50000, value=5000)
            with col4b:
                groceries_utilities = st.number_input("Groceries & Utilities (â‚¹)", min_value=0, max_value=100000, value=15000)
                other_monthly_expenses = st.number_input("Other Monthly Expenses (â‚¹)", min_value=0, max_value=100000, value=5000)

        # Loan Request Details
        with st.expander("ðŸ¦ Loan Request Details", expanded=False):
            col5a, col5b = st.columns(2)
            with col5a:
                requested_amount = st.number_input("Requested Loan Amount (â‚¹)", min_value=0, max_value=5000000, value=500000, step=10000)
                requested_tenure = st.slider("Requested Tenure (Months)", min_value=6, max_value=84, value=36)
            with col5b:
                emi_scenario = st.selectbox(
                    "Loan Purpose",
                    [
                        "Personal Loan EMI",
                        "Vehicle EMI",
                        "Home Appliances EMI",
                        "Education EMI",
                        "E-commerce Shopping EMI",
                    ],
                )
                existing_loans = st.radio("Existing Loans", ["No", "Yes"])

    with col2:
        st.subheader("EMI Prediction")

        total_expenses = (
            monthly_rent
            + school_fees
            + college_fees
            + travel_expenses
            + groceries_utilities
            + other_monthly_expenses
            + current_emi_amount
        )
        disposable_income = monthly_salary - total_expenses
        savings_ratio = (bank_balance / monthly_salary) if monthly_salary > 0 else 0
        debt_to_income = (current_emi_amount / monthly_salary) if monthly_salary > 0 else 0

        st.markdown("### ðŸ“ˆ Financial Health")
        # Disposable income
        st.success(f"Disposable Income: â‚¹{disposable_income:,.0f}") if disposable_income > 0 else st.error(
            f"Disposable Income: â‚¹{disposable_income:,.0f}"
        )
        # Savings ratio
        if savings_ratio >= 6:
            st.success(f"Savings Ratio: {savings_ratio:.1f} months")
        elif savings_ratio >= 3:
            st.warning(f"Savings Ratio: {savings_ratio:.1f} months")
        else:
            st.error(f"Savings Ratio: {savings_ratio:.1f} months")
        # DTI
        if debt_to_income <= 0.3:
            st.success(f"Debt-to-Income: {debt_to_income:.1%}")
        elif debt_to_income <= 0.5:
            st.warning(f"Debt-to-Income: {debt_to_income:.1%}")
        else:
            st.error(f"Debt-to-Income: {debt_to_income:.1%}")
        # Credit score
        if credit_score >= 750:
            st.success(f"Credit Score: {credit_score} (Excellent)")
        elif credit_score >= 700:
            st.info(f"Credit Score: {credit_score} (Good)")
        elif credit_score >= 650:
            st.warning(f"Credit Score: {credit_score} (Fair)")
        else:
            st.error(f"Credit Score: {credit_score} (Poor)")

        st.markdown("---")
        if st.button("ðŸš€ Predict My EMI", use_container_width=True):
            with st.spinner("Analyzing your financial profile..."):
                input_data = {
                    "age": age,
                    "monthly_salary": monthly_salary,
                    "years_of_employment": years_of_employment,
                    "monthly_rent": monthly_rent,
                    "family_size": family_size,
                    "dependents": dependents,
                    "school_fees": school_fees,
                    "college_fees": college_fees,
                    "travel_expenses": travel_expenses,
                    "groceries_utilities": groceries_utilities,
                    "other_monthly_expenses": other_monthly_expenses,
                    "current_emi_amount": current_emi_amount,
                    "credit_score": credit_score,
                    "bank_balance": bank_balance,
                    "emergency_fund": emergency_fund,
                    "requested_amount": requested_amount,
                    "requested_tenure": requested_tenure,
                    # categoricals
                    "marital_status": marital_status,
                    "education": education,
                    "employment_type": employment_type,
                    "company_type": company_type,
                    "emi_scenario": emi_scenario,
                    "existing_loans": existing_loans,
                }
                predicted_emi = predictor.predict_emi(input_data)
                if predicted_emi is not None:
                    display_prediction_results(predicted_emi, disposable_income, requested_amount)


# ------------------------------
# Results box
# ------------------------------

def display_prediction_results(predicted_emi: float, disposable_income: float, requested_amount: float):
    st.markdown('<div class="prediction-box">', unsafe_allow_html=True)
    st.markdown("### ðŸŽ¯ EMI Prediction Result")

    baseline = (disposable_income * 0.5) if disposable_income > 0 else 0
    delta_val = predicted_emi - baseline

    st.metric(label="Maximum Affordable EMI", value=f"â‚¹{predicted_emi:,.0f}", delta=f"{delta_val:,.0f}")

    affordability_ratio = (predicted_emi / disposable_income) if disposable_income > 0 else np.inf
    if disposable_income <= 0:
        st.error("âŒ Disposable income is non-positive. Consider reducing expenses or increasing income.")
    else:
        if affordability_ratio <= 0.4:
            st.success("âœ… **Excellent Affordability** - Well within your means")
        elif affordability_ratio <= 0.6:
            st.info("â„¹ï¸ **Good Affordability** - Manageable with your income")
        elif affordability_ratio <= 0.8:
            st.warning("âš ï¸ **Moderate Affordability** - Consider reducing loan amount")
        else:
            st.error("âŒ **High Risk** - May strain your finances")

    st.markdown("#### ðŸ’¡ Recommendations:")
    if predicted_emi < 5000:
        st.write("- Consider improving your credit score")
        st.write("- Build emergency savings")
        st.write("- Explore smaller loan amounts")
    elif predicted_emi < 15000:
        st.write("- Your financial profile supports moderate borrowing")
        st.write("- Maintain current debt levels")
        st.write("- Continue building savings")
    else:
        st.write("- Strong financial position for larger loans")
        st.write("- Consider investment opportunities")
        st.write("- Maintain financial discipline")

    st.markdown("</div>", unsafe_allow_html=True)


# ------------------------------
# Financial analysis tab
# ------------------------------

def display_financial_analysis():
    st.subheader("ðŸ“ˆ Financial Analysis & Insights")

    col1, col2 = st.columns(2)
    with col1:
        st.markdown("### Key Factors Affecting EMI")
        factors = {
            "Credit Score": "Higher scores increase borrowing capacity",
            "Monthly Income": "Primary determinant of affordable EMI",
            "Existing Debt": "Reduces available EMI capacity",
            "Employment Stability": "Longer tenure improves eligibility",
            "Savings Ratio": "Indicates financial discipline",
            "Expense Ratio": "Lower ratios mean higher disposable income",
        }
        for factor, impact in factors.items():
            with st.expander(f"ðŸ“Š {factor}"):
                st.write(impact)

    with col2:
        st.markdown("### ðŸ’¡ Tips to Improve EMI Eligibility")
        tips = [
            "Maintain credit score above 750 for best rates",
            "Keep debt-to-income ratio below 40%",
            "Build emergency fund of 3-6 months expenses",
            "Show stable employment history (2+ years)",
            "Reduce unnecessary monthly expenses",
            "Clear existing high-interest debts first",
        ]
        for i, tip in enumerate(tips, 1):
            st.write(f"{i}. {tip}")

    # Quick EMI Calculator
    st.markdown("### ðŸ§® Quick EMI Calculator")
    col_calc1, col_calc2, col_calc3 = st.columns(3)
    with col_calc1:
        loan_amount = st.number_input("Loan Amount (â‚¹)", min_value=1000, max_value=10000000, value=500000, step=10000)
    with col_calc2:
        interest_rate = st.slider("Interest Rate (%)", min_value=5.0, max_value=20.0, value=10.5, step=0.1)
    with col_calc3:
        tenure_years = st.slider("Tenure (Years)", min_value=1, max_value=30, value=5)

    if st.button("Calculate EMI"):
        tenure_months = int(tenure_years * 12)
        monthly_rate = float(interest_rate) / 12 / 100
        if monthly_rate > 0:
            emi = (loan_amount * monthly_rate * (1 + monthly_rate) ** tenure_months) / (
                (1 + monthly_rate) ** tenure_months - 1
            )
            total_payment = emi * tenure_months
            total_interest = total_payment - loan_amount
            st.success(f"**Monthly EMI:** â‚¹{emi:,.0f}")
            st.info(f"**Total Interest:** â‚¹{total_interest:,.0f}")
            st.info(f"**Total Payment:** â‚¹{total_payment:,.0f}")


# ------------------------------
# About tab
# ------------------------------

def display_about_section():
    st.subheader("â„¹ï¸ About EMI Amount Predictor")
    st.markdown(
        """
        ### How It Works

        The EMI Amount Predictor uses machine learning to analyze your financial profile and predict
        the maximum affordable EMI amount you can comfortably manage based on:

        - **Personal Information**: Age, marital status, dependents
        - **Employment Details**: Income, job stability, company type
        - **Financial Health**: Credit score, savings, existing debts
        - **Monthly Expenses**: Living costs, education, utilities
        - **Loan Requirements**: Amount, tenure, purpose

        ### ðŸŽ¯ Model Accuracy (Illustrative)

        - **RÂ² Score**: ~0.85+ (varies by dataset)
        - **Average Error**: ~Â±â‚¹1,500
        - **Accuracy**: ~90% within 20% of actual EMI

        *Figures are indicative; consult your experiment tracking for exact numbers.*

        ### âš ï¸ Important Notes

        - This is a predictive model and not financial advice
        - Actual loan approvals depend on lender policies
        - Maintain healthy financial habits for better eligibility
        - Regularly monitor your credit score

        ### ðŸ“ž Need Help?

        For financial advice, consult with a certified financial planner.
        For technical issues, contact our support team.
        """
    )


# ------------------------------
# Model info tab
# ------------------------------

def display_model_info(predictor: EMIPredictor):
    st.subheader("ðŸ”§ Model Information")

    col1, col2 = st.columns(2)
    with col1:
        st.markdown("### Model Details")
        st.write("**Algorithm:** XGBoost Regressor")
        st.write("**Training Data:** 1,000+ loan applications (example)")
        st.write("**Features Used:** Top financial indicators from profile & expenses")
        st.write("**Last Updated:** Recent")

        st.markdown("### Key Performance Metrics (Example)")
        st.metric("RÂ² Score", "0.87")
        st.metric("RMSE", "â‚¹1,423")
        st.metric("MAE", "â‚¹1,150")

    with col2:
        st.markdown("### Top 10 Influential Features")
        # Prefer dynamic importances if available; else fallback
        shown_dynamic = False
        try:
            if hasattr(predictor.model, "feature_importances_") and predictor.feature_names is not None:
                importances = np.array(predictor.model.feature_importances_, dtype=float)
                names = np.array(predictor.feature_names)
                order = np.argsort(importances)[::-1][:10]
                for name, imp in zip(names[order], importances[order]):
                    st.progress(float(imp), text=f"{name}: {imp:.0%}")
                shown_dynamic = True
        except Exception:
            pass

        if not shown_dynamic:
            feature_importance_example = {
                "Monthly Salary": 0.18,
                "Credit Score": 0.15,
                "Disposable Income": 0.12,
                "Years of Employment": 0.09,
                "Bank Balance": 0.08,
                "Current EMI Obligations": 0.07,
                "Debt-to-Income Ratio": 0.06,
                "Employment Type": 0.05,
                "Emergency Fund": 0.04,
                "Age": 0.03,
            }
            for feature, importance in feature_importance_example.items():
                st.progress(importance, text=f"{feature}: {importance:.0%}")

    st.markdown("### ðŸ“Š Model Validation (Typical)")
    st.write(
        "\n".join(
            [
                "- Cross-validation (5-fold)",
                "- Out-of-sample testing",
                "- Real-world performance monitoring",
                "- Regular updates with new data",
            ]
        )
    )


# ------------------------------
# Entrypoint
# ------------------------------
if __name__ == "__main__":
    main()
```
